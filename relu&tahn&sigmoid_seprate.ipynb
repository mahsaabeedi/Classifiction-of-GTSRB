```python
import cv2
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


dim = (30, 30)
X_train =[]
Y_train=[]
n=0

#read train data
path='/kaggle/input/gtsrb-german-traffic-sign/train'
grup = os.listdir(path)
for i in grup:
    tem=os.path.join(path, i)
    img_nams=os.listdir(tem)
    n+=1
    for imasg in img_nams:
        img=np.array(Image.open(os.path.join(tem,imasg )).resize((30,30),Image.ANTIALIAS))
        X_train.append(img)
        s=int(i)
        Y_train.append(s)



X_t=np.array(X_train)
X_t = X_t.astype('float32')
X_train=X_t/255.0


Y_t=np.array(Y_train)
from keras.utils import np_utils
Y_train=np_utils.to_categorical(Y_t)
#print(n)

#SHUFFLE DATA
from sklearn.utils import shuffle
X_train, Y_train = shuffle(X_train,Y_train, random_state=0)


```

```python
# relu model
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout

model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(5,5) ,activation='relu', input_shape=X_train.shape[1:]))
model.add(Conv2D(filters=64, kernel_size=(3, 3),strides =2 , activation='relu'))
#model.add(MaxPool2D(pool_size=(2, 2)))
# model.add(Dropout(rate=0.25))
model.add(Conv2D(filters=64, kernel_size=(3, 3),strides =2, activation='relu'))
#model.add(MaxPool2D(pool_size=(2, 2)))
# model.add(Dropout(rate=0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
# model.add(Dropout(rate=0.5))
model.add(Dense(43, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


trained_model = model.fit(X_train, Y_train, batch_size=128, epochs=15 , validation_split = 0.2)

history=trained_model.history

#plot _relu

import matplotlib.pyplot as plt

losses_relu=history['loss']
val_losses_relu=history['val_loss']

acc_relu=history['accuracy']
val_acc_relu = history['val_accuracy']


plt.figure(figsize=(10,10))
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(losses_relu)
plt.plot(val_losses_relu)
plt.legend(['train loss','val_loss'])
plt.title('Relu')

plt.figure(figsize=(10,10))
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.plot(acc_relu)
plt.plot(val_acc_relu)
plt.legend(['train accuracy','val_accuracy'])
plt.title('Relu')

#test acc & loss relu
```

```python
# tanh model1
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout

model1 = Sequential()
model1.add(Conv2D(filters=32, kernel_size=(5,5) ,activation='tanh', input_shape=X_train.shape[1:]))
model1.add(Conv2D(filters=64, kernel_size=(3, 3),strides =2 , activation='tanh'))
#model.add(MaxPool2D(pool_size=(2, 2)))
# model.add(Dropout(rate=0.25))
model1.add(Conv2D(filters=64, kernel_size=(3, 3),strides =2, activation='tanh'))
#model.add(MaxPool2D(pool_size=(2, 2)))
# model.add(Dropout(rate=0.25))
model1.add(Flatten())
model1.add(Dense(256, activation='tanh'))
# model.add(Dropout(rate=0.5))
model1.add(Dense(43, activation='softmax'))

model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

trained_model1 = model1.fit(X_train, Y_train, batch_size=128, epochs=15 , validation_split = 0.2)
history1=trained_model1.history

#plot _tanh
losses_tanh=history1['loss']
val_losses_tanh=history1['val_loss']

acc_tanh=history1['accuracy']
val_acc_tanh = history1['val_accuracy']



plt.figure(figsize=(10,10))
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(losses_tanh)
plt.plot(val_losses_tanh)
plt.legend(['train loss','val_loss'])
plt.title('tanh')

plt.figure(figsize=(10,10))
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.plot(acc_tanh)
plt.plot(val_acc_tanh)
plt.legend(['train accuracy','val_accuracy'])
plt.title('tanh')
```
